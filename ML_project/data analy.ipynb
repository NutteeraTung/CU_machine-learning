```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load data
df = pd.read_excel(r'D:\M.eng\Machine learning\ML_project\WaterSafety_Data.xlsx')

# Selecting X (first 20 columns) and Y (21st column)
X = df.iloc[:, 0:20]  # Features
y = df.iloc[:, 20]    # Target variable

# Add target variable to the dataset
df_plot = X.copy()
df_plot["Target"] = y  
# Create pairplot
sns.pairplot(df_plot, y_vars=["Target"], x_vars=df_plot.columns[:-1], kind="reg")
plt.show()
```
