```python
# Import necessary modules
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns


# Compute correlation matrix
co_mtx = df.corr(numeric_only=True)

# Print correlation matrix
print(co_mtx)

fig, ax = plt.subplots(figsize=(20, 10))  # Set the figure size
# Plot correlation heatmap
sns.heatmap(co_mtx, cmap="YlGnBu", annot=True)

# Display heatmap
plt.show()
```

```python
#drop if correlation is negative for improving models
prun_data = []
for column in corr_data.columns:
    # print(column)
    # print(corr_data.iloc[0][column])
    if abs(corr_data.iloc[0][column])<0.10:
        prun_data.append(column)
prun_df = corr_data.drop(prun_data,axis=1)
print(prun_df)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load data
df = pd.read_excel(r'D:\M.eng\Machine learning\ML_project\WaterSafety_Data.xlsx')

# Selecting X (first 20 columns) and Y (21st column)
X = df.iloc[:, 0:20]  # Features
y = df.iloc[:, 20]    # Target variable

# Add target variable to the dataset
df_plot = X.copy()
df_plot["Target"] = y  
# Create pairplot
sns.pairplot(df_plot, y_vars=["Target"], x_vars=df_plot.columns[:-1], kind="reg")
plt.show()
```
